```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(420)
```

```python
fastball_speed = np.random.randint(90, 106, size=500)
```

```python
tommy_john = np.where(fastball_speed > 96, np.random.choice([0, 1], size=500, p=[0.3, 0.7]), 0)
```

```python
d = {'fastball_speed': fastball_speed, 'tommy_john': tommy_john}
```

```python
df = pd.DataFrame(data = d)
```

```python
df.head()
```

```python
X = df[['fastball_speed']]
```

```python
y = df[['tommy_john']]
```

## **_Normal Train Test Split_**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=11)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train, y_train.values.ravel())
```

```python
lr.score(X_test, y_test)
```

## **_Test this again_**

```python
X2_train, X2_test, y2_train, y2_test = train_test_split(X, y, test_size=0.2, random_state=25)
```

```python
lr.fit(X2_train, y2_train.values.ravel())
```

```python
lr.score(X2_test, y2_test)
```

## **_Basic Cross Val_**

```python
from sklearn.model_selection import cross_val_score
```

```python
cv = cross_val_score(lr, X, y.values.ravel(), cv=10)
```

```python
cv
```

```python
np.average(cv)
```

```python
np.std(cv)
```

### _**Define Standard Kfold**_

```python
from sklearn.model_selection import KFold
```

```python
kf = KFold(n_splits=15, shuffle=True, random_state=42)
```

```python
kfscore = cross_val_score(lr, X, y.values.ravel(), cv=kf, scoring='f1')
```

```python
kfscore2 = cross_val_score(lr, X, y.values.ravel(), cv=kf, scoring='accuracy')
```

```python
kfscore
```

```python
np.average(kfscore)
```

```python
np.std(kfscore)
```

```python
kfscore2
```

```python
np.average(kfscore2)
```

```python
np.std(kfscore2)
```

## **_Startified K Fold_**

```python
from sklearn.model_selection import StratifiedKFold
```

```python
kfs = StratifiedKFold(n_splits=10, shuffle=True, random_state=11)
```

```python
kfsscore =cross_val_score(lr, X, y.values.ravel(), cv=kfs)
```

## **_Pipeline Example_**

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
from sklearn.pipeline import make_pipeline
```

```python
pipe = make_pipeline(scaler, lr)
```

```python
pipe.fit(X_train, y_train.values.ravel())
```

```python
scorepipe = cross_val_score(pipe, X, y.values.ravel(), cv=10)
scorepipe
```

```python
np.average(scorepipe)
```

```python
np.std(scorepipe)
```
