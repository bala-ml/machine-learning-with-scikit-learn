```python
import numpy as np
import pandas as pd
```

```python
mean = 55
std_dev = 10
num_samples = 500

column1_numbers = np.random.normal(mean, std_dev, num_samples)
column1_numbers = np.clip(column1_numbers, 30, 120)
column1_numbers = np.round(column1_numbers).astype(int)

mean2 = 18
std_dev2 = 3

column2_numbers = np.random.normal(mean2, std_dev2, num_samples)
column2_numbers = np.clip(column2_numbers, 12, 26)
column2_numbers = np.round(column2_numbers).astype(int)

column3_numbers = np.random.randint(2, size=num_samples)
column3_numbers[column1_numbers > mean] = 1

data = {'Miles_Per_week': column1_numbers,
        'Farthest_run': column2_numbers,
        'Qualified_Boston_Marathon': column3_numbers}

df = pd.DataFrame(data)
```

```python
df.head()
```

```python
X = df.iloc[:, 0:2]
```

```python
y = df.iloc[:, 2]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=26)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
```

```python
param_grid = [{
    'n_estimators': [500, 1000, 1500],
    # 'criterion': ['entropy', 'gini'],
    'min_samples_split': [5,10, 15],
    'min_samples_leaf': [1, 2, 4]
    # 'max_depth': [10, 20, 30]
}]
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
gr = GridSearchCV(rfc, param_grid, cv=2, scoring='accuracy', n_jobs=-1)
```

```python
gr.fit(X_train, y_train)
```

```python
gr.best_score_
```

```python
gr.best_params_
```

## **_Randomized_**

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
random_param_grid = [{
    'n_estimators': [500, 1000, 1500],
    'criterion': ['entropy', 'gini'],
    'min_samples_split': [5,10, 15],
    'min_samples_leaf': [1, 2, 4],
   'max_depth': [10, 20, 30]
}]
```

```python
rgr = RandomizedSearchCV(rfc, random_param_grid, cv=5, scoring='accuracy', n_jobs=-1, random_state=26)
```

```python
rgr.fit(X_train, y_train)
```

```python
rgr.best_score_
```
