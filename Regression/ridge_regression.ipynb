# **_Ridge Regression_**

```python
from sklearn.datasets import make_regression

from Regression.multiple_linear_regression import X_train, X_test, y_train, y_test
```

```python
X, y = make_regression(n_samples=100, n_features=4, noise=0.1, random_state=42, effective_rank=2)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=19)
```

# **_Before applying Ridge Regression, the data should be standardized_**

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
from sklearn.linear_model import Ridge
```

```python
ridge = Ridge()
```

```python
ridge.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
y_pred = ridge.predict(X_test)
```

```python
mean_squared_error(y_test, y_pred)
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
r2_score(y_test, y_pred)
```

# **_Hyperparameter grid used for model tuning_**

```python
param_grid = {
    'alpha' : [0.001, 0.01, 0.1, 1.0, 10.0, 100.0]
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
ridge_cv =GridSearchCV(ridge, param_grid=param_grid, cv=3, n_jobs=-1)
```

```python
ridge_cv.fit(X_train, y_train)
```

```python
y_pred2 = ridge_cv.predict(X_test)
```

```python
mean_absolute_error(y_test, y_pred2)
```

```python
mean_squared_error(y_test, y_pred2)
```

```python
r2_score(y_test, y_pred2)
```
