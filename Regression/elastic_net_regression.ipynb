```python
import pandas as pd

from Regression.multiple_linear_regression import X_train, X_test
```

```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.shape
```

```python
tips.info()
```

```python
tips = pd.get_dummies(tips, dtype=int)
```

```python
tips.head()
```

```python
X = tips.drop('tip', axis=1)
```

```python
y = tips['tip']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=19)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.fit_transform(X_test)
```

```python
from sklearn.linear_model import ElasticNet
```

```python
elastic_net = ElasticNet()
```

```python
elastic_net.fit(X_train, y_train)
```

```python
y_pred = elastic_net.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
mean_squared_error(y_test, y_pred)
```

```python
r2_score(y_test, y_pred)
```

```python
param_grid = { "alpha": [0.1, 0.3, 0.5, 0.7, 0.9],
               'l1_ratio': [0.1, 0.3, 0.5, 0.7, 0.9],}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
elastic_net_cv = GridSearchCV(elastic_net, param_grid, cv=3, n_jobs=-1, scoring='neg_mean_squared_error')
```

```python
elastic_net_cv.fit(X_train, y_train)
```

```python
y_pred2 = elastic_net_cv.predict(X_test)
```

```python
mean_squared_error(y_test, y_pred2)
```

```python
mean_absolute_error(y_test, y_pred2)
```

```python
r2_score(y_test, y_pred2)
```
