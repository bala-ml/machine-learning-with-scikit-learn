```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

from Classification.logistic_regression import X_train, X_test, y_train
```

```python
iris = load_iris()
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
df['target'] = iris.target
```

```python
df.head()
```

```python
iris.target_names
```

```python
df['flower_name'] = df.target.apply(lambda x: iris.target_names[x])
```

```python
df['flower_name'].unique()
```

```python
setosa_df     = df[df.target == 0]
versicolor_df = df[df.target == 1]
virginica_df  = df[df.target == 2]
```

```python
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.scatter(setosa_df['sepal length (cm)'], setosa_df['sepal width (cm)'],color="green",marker='x')
plt.scatter(versicolor_df['sepal length (cm)'], versicolor_df['sepal width (cm)'],color="blue",marker='o')
```

```python
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.scatter(setosa_df['petal length (cm)'], setosa_df['petal width (cm)'],color="green",marker='x')
plt.scatter(versicolor_df['petal length (cm)'], versicolor_df['petal width (cm)'],color="blue",marker='o')
```

```python
X = df.drop(['target','flower_name'], axis='columns')
```

```python
y = df.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.svm import SVC
```

```python
model = SVC()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
model.score(X_test, y_test)
```

## **_Regularization_**

```python
model_C = SVC(C=1)
model_C.fit(X_train, y_train)
model_C.score(X_test, y_test)
```

```python
model_C = SVC(C=10)
model_C.fit(X_train, y_train)
model_C.score(X_test, y_test)
```

## **_Influence of points_**

```python
model_g = SVC(gamma=10)
model_g.fit(X_train, y_train)
model_g.score(X_test, y_test)
```

## **_Type of boundary_**

```python
model_linear_kernal = SVC(kernel='linear')
model_linear_kernal.fit(X_train, y_train)
model_linear_kernal.score(X_test, y_test)
```
