```python
import pandas as pd

from Classification.logistic_regression import X_train, y_train
```

```python
df = pd.read_csv('../Data/pearl_jam_tour.csv')
```

```python
df.head()
```

```python
df2 = pd.get_dummies(df[['Continent', 'Day of Week']], dtype=int)
```

```python
df2.head()
```

```python
df3 = pd.concat([df, df2], axis=1)
```

```python
df4 = df3.drop(['Continent', 'Day of Week'], axis=1)
```

```python
df4.columns
```

```python
X = df4.drop(['Sold Out'], axis=1)
```

```python
y = df4['Sold Out']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()
```

```python
gnb.fit(X_train, y_train)
```

```python
y_pred = gnb.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
cr = classification_report(y_test, y_pred)
```

```python
print(cr)
```

```python
gnb.score(X_train, y_train)
```

```python
gnb.score(X_test, y_test)
```
