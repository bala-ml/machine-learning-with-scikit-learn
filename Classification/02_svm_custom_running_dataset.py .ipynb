```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
mean = 55
std_dev = 10
num_samples = 500

column1_num = np.random.normal(mean, std_dev, num_samples)
column1_num = np.clip(column1_num, 30, 120)
column1_num = np.round(column1_num).astype(int)

mean2 = 18
std_dev2 = 3

column2_num = np.random.normal(mean2, std_dev2, num_samples)
column2_num = np.clip(column2_num, 12, 26)
column2_num = np.round(column2_num).astype(int)

column3_num = np.random.randint(2, size=num_samples)
column3_num[column1_num > mean] = 1

data = {
    'miles_per_week': column1_num,
    'farthest_run':  column2_num,
    'qualified_boston_marathon': column3_num
}
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['miles_per_week'], df['farthest_run'], c=df['qualified_boston_marathon'], cmap='coolwarm')
plt.xlabel('Miles per Week')
plt.ylabel('Farthest Run')
plt.title('Scatter Plot')
plt.colorbar(label='qualified_boston_marathon')
plt.show()
```

```python
X = df.iloc[:, 0:2]
```

```python
y = df.iloc[:, 2]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split = train_test_split(X, y, test_size=0.2, random_state=33)
```

```python
from sklearn.svm import SVC
```

```python
model = SVC()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

## **_Regularization_**

```python
model_reg0 = SVC(C=0.1)
```

```python
model_reg0.fit(X_train, y_train)
```

```python
model_reg0.score(X_test, y_test)
```

```python
model_reg1 = SVC(C=1)
```

```python
model_reg1.fit(X_train, y_train)
```

```python
model_reg1.score(X_test, y_test)
```
