* **_G – Games played_**

* **_R – Runs scored_**
* **_H – Hits_**
* **_2B – Doubles (2-base hits)_**
* **_3B – Triples (3-base hits)_**
* **_HR – Home runs_**
* **_RBI – Runs Batted In (runs scored because of the player’s hit)_**
* **_BB – Walks (Base on Balls)_**
* **_SO – Strikeouts_**
* **_SB – Stolen bases_**
* **_CS – Caught stealing_**
* **_BA – Batting Average (Hits ÷ At-bats)_**
* **_HOF – Hall of Fame status_**

# **_Find out - Hall of Fame status_**

```python
X = df.drop(columns=['PLAYER', 'HOF'])
```

```python
y = df['HOF']
```

```python
X.head()
```

```python
X.shape
```

```python
y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2)
# random_state fixes the randomness so that the same random sample is generated in every iteration.
# test_size decides how much data goes to the test
```
