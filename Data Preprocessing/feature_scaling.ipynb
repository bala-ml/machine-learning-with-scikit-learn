# **_Feature Scaling_**

# **_Normalization (Minâ€“Max Scaling) - Range lies between [0, 1] or [-1, 1]._**

##     **_formula = X - min(X) / max(X) - min(X)_**


# **_Standardization (Z-score Scaling) - mean 0, std 1, no fixed bounds._**

```python
import pandas as pd
```

```python
df = pd.read_csv('../Data/500hits.csv', encoding='latin-1')
```

```python
df.head()
```

```python
df = df.drop(columns=['PLAYER', 'CS'])
```

```python
df.info()
```

```python
df.describe().round()
```

```python
X1 = df.iloc[:, 0:13] # rows, columns [0:13 or : all columns]
```

```python
X2 = df.iloc[:, 0:13]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaleStandard = StandardScaler()
```

```python
X1 = scaleStandard.fit_transform(X1)
```

```python
X1 = pd.DataFrame(X1, columns=['YRS', 'G', 'AB', 'R', 'H', '2B', '3B', 'HR', 'RBI', 'BB', 'SO', 'SB', 'BA'])
# Column names are reassigned because sklearn transformers return NumPy arrays, which do not retain column names.
```

```python
X1.head()
```

```python
X1.describe().round().T
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaleMinMax = MinMaxScaler(feature_range=(0, 1))
```

```python
X2 = scaleMinMax.fit_transform(X2)
```

```python
X2 = pd.DataFrame(X2, columns=['YRS', 'G', 'AB', 'R', 'H', '2B', '3B', 'HR', 'RBI', 'BB', 'SO', 'SB', 'BA'])
```

```python
X2.head()
```

```python
X2.describe().round()
```
