# **_One-Hot-Encoder - Categorical Data_**

```python
import pandas as pd
```

```python
d = {'sales': [100000,222000,1000000,522000,111111,222222,1111111,20000,75000,90000,1000000,10000],
'city': ['Tampa','Tampa','Orlando','Jacksonville','Miami','Jacksonville','Miami','Miami','Orlando','Orlando','Orlando','Orlando'],
'size': ['Small', 'Medium','Large','Large','Small','Medium','Large','Small','Medium','Medium','Medium','Small',]}
```

```python
df = pd.DataFrame(data = d)
```

```python
df.head()
```

```python
df['city'].nunique()
```

```python
df['city'].unique()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(handle_unknown='ignore', sparse_output=False).set_output(transform='pandas')
# handle_unknown='ignore' - if new category appears in test data , sets all zeros for that row
# sparse[stores them in a compressed format]: Dense (sparse_output=False) - Attendance sheet with Present / Absent for everyone , Sparse (sparse_output=True) - Sheet that records only who is Present
# .set_output(transform='pandas') - Forces output to be a pandas DataFrame
```

```python
ohetransform = ohe.fit_transform(df[['city']])
```

```python
ohetransform
```

```python
df.head()
```

```python
ohetransform.head()
```

```python
df = pd.concat([df, ohetransform], axis=1).drop(columns=['city'])
```
