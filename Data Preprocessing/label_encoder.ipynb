# **_Label Encoder_**

```python
import pandas as pd
```

```python
d = {'gender': ['Male', 'Female', 'Male', 'Female', 'Male']}
```

```python
df = pd.DataFrame(data=d)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
enc_gender = le.fit_transform(df['gender'])
```

```python
enc_gender
```

```python
le.classes_
```

```python
df['gender'] = enc_gender
```
